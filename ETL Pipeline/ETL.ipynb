```python
import requests
import pandas as pd

url = "https://dogapi.dog/api/v2/breeds"
response = requests.get(url)

data = response.json()

clean_data = []

for breed in data["data"]:
    attr = breed.get("attributes", {})

    clean_data.append({
        "name": attr.get("name", None),
        "description": attr.get("description", None),
        "life_span": str(attr.get("life", None))  # convert to string if it's nested
    })

df = pd.DataFrame(clean_data)

# Basic cleaning
df.drop_duplicates(inplace=True)
df.fillna("Unknown", inplace=True)

df.head()
```
