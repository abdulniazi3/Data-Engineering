```python
import requests

api_key = "b8740dd60a554754bd2796a720decf3c"

url = "https://newsapi.org/v2/everything"

params = {
    "q": "Barcelona",
    "language": "en",
    "pageSize": 10,
    "apiKey": api_key
}

response = requests.get(url, params=params)

data = response.json()

print(data)
```

```python
articles = []

for article in data["articles"]:
    articles.append({
        "title": article["title"],
        "source": article["source"]["name"],
        "published": article["publishedAt"],
        "description": article["description"]
    })

import pandas as pd
df = pd.DataFrame(articles)

df.head()
```
